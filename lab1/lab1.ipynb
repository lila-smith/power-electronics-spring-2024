Q12: Create a plot for the voltage conversion ratio as a function of inductance (D=10%) for L=1 uH to 15 uH, taking 1 uH increments.

```python
import numpy as np
import plotly.express as px
input_voltage = 100 #V
inductance = np.linspace(1,15, 15) # uH
voltage_out = [27.2691, 20.1130, 16.7363, 14.6591, 13.2132,
               12.1309, 11.2808, 10.5899, 10.0137, 9.9990,
               9.9990, 9.9990, 9.9990, 9.9990, 9.9990] #V, average
voltage_conversion = np.array(voltage_out)/input_voltage
fig = px.scatter(x=inductance, y=voltage_conversion, title="Inductance and Voltage Conversion Ratio")
fig.update_layout(
    xaxis_title="Inductance (mH)",
    yaxis_title="Voltage Conversion Ratio"
)
fig.show()
```

Q16: Plot the voltage conversion ratio M as a function of the duty cycle, for D=10% to 90% in 10% increments.

```python
import numpy as np
import plotly.express as px
input_voltage = 100 #V
duty_cycle = np.linspace(10,90, 9) # %
voltage_out = [123.4308, 156.1538, 203.8359,
               277.2158, 398.7048, 621.6851,
               1110.3874, 2410.9605, 6113.1515] #V, average
voltage_conversion = np.array(voltage_out)/input_voltage
fig = px.scatter(x=duty_cycle,
                 y=voltage_conversion,
                 title="Inductance and Voltage Conversion Ratio")
fig.update_layout(
    xaxis_title="Duty Cycle (%)",
    yaxis_title="M: Voltage Conversion Ratio"
)
fig.show()
```
